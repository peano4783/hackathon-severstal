```python
def collect(array):
    answer = []
    for a in array:
        if not (a is None):
            for it in a:
                answer.append(it)
    if len(answer)== 0:
        answer = None
    return answer

matrix["total"] = matrix.apply(lambda x: collect(x), axis=1) 
matrix["total"].to_csv("dict_short.csv")
```
